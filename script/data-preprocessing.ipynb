```python
#Gerekli kütüphaneler
import pandas as pd
import numpy as np
import random
```

```python
#RNA-Seq veriseti dosyası okunup sütun adı değiştirildi
gedf = pd.read_csv("../data/GSE54456_RPKM_samples.txt", sep="\t")
gedf.rename(columns={"Unnamed: 0" : "gene"}, inplace=True)
gedf
```

```python
#TF-Target veriseti okunup birleştirme için ortak sütun belirlendi.
ttdf = pd.read_csv("../data/tf-target.txt")
ttdf.rename(columns={"tf":"gene"}, inplace=True)
ttdf
```

```python
#TF-Target çiftleri için RNA-seq değeri eşleştirildi
mdf = ttdf.merge(gedf)
#Gerekli olmayan sütunlar düşürüldü
genel_df2 = mdf.drop(["references_PMID","mode_of_regulation"], axis=1)
#Bu çiftler için yeni bir class oluşturuldu "pair" sütunu eğitimde bizim etiketimiz olacak
genel_df2["pair"] = 1
genel_df2
```

```python
#Rastgele tf-target çiftleri oluşturuldu.
lst = gedf["gene"]
lst = list(lst)
def pop_random(lst):
    idx = random.randrange(0, len(lst))
    return lst.pop(idx)
pairs = []
while lst:
    rand1 = pop_random(lst)
    rand2 = pop_random(lst)
    pair = rand1, rand2
    pairs.append(pair)
```

```python
generated_df = pd.DataFrame(pairs)
generated_df.rename(columns={0:"gene",1:"target"}, inplace=True)
#Bilinen TF-Target çiftleri ile eşit sayıda olması için oluşan dataframe sınırlandırıldı.
short_gdf = generated_df.iloc[:9378,:]
#Rastgele oluşturulan TF-Target çiftleri için RNA-seq değerleri eşleştirildi
false_shortdf = short_gdf.merge(gedf)
#Bu çiftler için class'ımız olan "pair" sütunu 0 olarak atandı.
false_shortdf["pair"] = 0
false_shortdf
```
